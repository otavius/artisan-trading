```python
import yfinance as yf 
import pandas as pd 
import numpy as np 
from pandas.tseries.offsets import DateOffset
```

```python
df_ = yf.download("AAPL",start="2024-01-01", auto_adjust=False)
```

```python
vix_df = yf.download("^VIX", start="1990-01-01", auto_adjust=False)
```

```python
vix_df = vix_df.xs("^VIX", axis=1, level="Ticker")
```

```python
vix_df
```

```python
vix_df["MA"] = vix_df.Close.rolling("30D").mean()
```

```python
vix_df
```

```python
vix_df_filt = vix_df[vix_df.Close > 1.5 * vix_df.MA]
```

```python
vix_df_filt
```

```python
series = pd.Series(vix_df_filt.index).diff() / np.timedelta64(1, "D") >= 30
```

```python
series[0] = True
```

```python
signals = vix_df_filt[series.values]
```

```python
signals.shape
```

```python
signals
```

```python
sp_df = yf.download("^GSPC", start="1990-01-01", auto_adjust=False)
sp_df = sp_df.xs("^GSPC", axis=1, level='Ticker')
```

```python
sp_df
```

```python
test_ = sp_df[(sp_df.index >= signals.index[0]) & (sp_df.index <=signals.index[0] + DateOffset(months=6))]
```

```python
(test_.Close.pct_change() + 1).cumprod()
```

```python
returns = []

for i in range(len(signals)):
    subdf = sp_df[(sp_df.index >= signals.index[i]) & (sp_df.index <=signals.index[i] + DateOffset(months=6))]
    returns.append((subdf.Close.pct_change() + 1).prod())
```

```python
pd.Series(returns).plot(kind="bar")
```
